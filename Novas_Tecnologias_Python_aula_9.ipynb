# Categorias de aprendizado de máquina

No nível mais fundamental, o aprendizado de máquina pode ser categorizado em dois tipos principais: aprendizado supervisionado e aprendizado não supervisionado.

- A *aprendizagem supervisionada* envolve, de alguma forma, a modelagem da relação entre as características medidas dos dados e algum rótulo associado aos dados; uma vez que este modelo é determinado, ele pode ser usado para aplicar rótulos a dados novos e desconhecidos. Isso é subdividido em tarefas de classificação e tarefas de regressão: na classificação, os rótulos são categorias discretas, enquanto na regressão, os rótulos são quantidades contínuas. Veremos exemplos de ambos os tipos de aprendizagem supervisionada na seção seguinte.

- O *aprendizado não supervisionado* envolve a modelagem dos recursos de um conjunto de dados sem referência a qualquer rótulo e é frequentemente descrito como "deixar o conjunto de dados falar por si mesmo". Esses modelos incluem tarefas como armazenamento em cluster e redução de dimensionalidade. Os algoritmos de agrupamento identificam grupos distintos de dados, enquanto os algoritmos de redução de dimensionalidade buscam representações mais sucintas dos dados. Veremos exemplos de ambos os tipos de aprendizagem não supervisionada na seção seguinte.

- Além disso, existem os chamados métodos de aprendizagem semissupervisionados, que ficam em algum lugar entre a aprendizagem supervisionada e a aprendizagem não supervisionada. Os métodos de aprendizagem semissupervisionados são frequentemente úteis quando apenas rótulos incompletos estão disponíveis.

# Classificação: previsão de rótulos discretos

Vamos primeiro dar uma olhada em uma tarefa de classificação simples, na qual você recebe um conjunto de pontos rotulados e deseja usá-los para classificar alguns pontos não rotulados.

## O código abaixo não é relevante nesse momento. Ele aparece aqui só para gerar a imagem que vem na sequência.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
import os
if not os.path.exists('figures'):
    os.makedirs('figures')
```

```python
# common plot formatting for below
def format_plot(ax, title):
    ax.xaxis.set_major_formatter(plt.NullFormatter())
    ax.yaxis.set_major_formatter(plt.NullFormatter())
    ax.set_xlabel('feature 1', color='gray')
    ax.set_ylabel('feature 2', color='gray')
    ax.set_title(title, color='gray')
```

```python
from sklearn.datasets import make_blobs
from sklearn.svm import SVC

# create 50 separable points
X, y = make_blobs(n_samples=50, centers=2,
                  random_state=0, cluster_std=0.60)

# fit the support vector classifier model
clf = SVC(kernel='linear')
clf.fit(X, y)

# create some new points to predict
X2, _ = make_blobs(n_samples=80, centers=2,
                   random_state=0, cluster_std=0.80)
X2 = X2[50:]

# predict the labels
y2 = clf.predict(X2)
```

```python
# plot the data
fig, ax = plt.subplots(figsize=(8, 6))
point_style = dict(cmap='Paired', s=50)
ax.scatter(X[:, 0], X[:, 1], c=y, **point_style)

# format plot
format_plot(ax, 'Input Data')
ax.axis([-1, 4, -2, 7])

#fig.savefig('figures/classificador1.png')
```

## Intuição geométrica da classificação

Imagine que temos os dados mostrados nesta figura acima.

Aqui temos dados bidimensionais: ou seja, temos duas dimensões (atributos ou features) para cada ponto, representados pelas posições (x, y) dos pontos no plano (de forma geral dizemos então que nossos dados existem num HIPERPLANO, que pode ter um número arbitrário de dimensões). Além disso, temos um de dois rótulos de classe para cada ponto, aqui representado pelas cores dos pontos. A partir desses recursos e rótulos, gostaríamos de criar um modelo que nos permitirá decidir se um novo ponto deve ser rotulado como "azul" ou "vermelho".

Existem vários modelos (hipóteses) possíveis para essa tarefa de classificação, mas aqui usaremos um extremamente simples. Faremos a suposição de que os dois grupos podem ser separados traçando uma linha reta através do plano entre eles, de modo que os pontos de cada lado da linha caiam no mesmo grupo. Aqui, o **modelo** é uma versão quantitativa da declaração "uma linha reta separa as classes", enquanto os **parâmetros do modelo** são os números particulares que descrevem a localização e orientação dessa linha para nossos dados. Os valores ideais para esses parâmetros do modelo são aprendidos a partir dos dados (esse é o "**aprendizado**" no aprendizado de máquina), que geralmente é chamado de treinamento do modelo.

A figura a seguir mostra uma representação visual da aparência do modelo treinado para esses dados:

```python
xx = np.linspace(-1, 4, 10)
yy = np.linspace(-2, 7, 10)
xy1, xy2 = np.meshgrid(xx, yy)
Z = np.array([clf.decision_function([t])
              for t in zip(xy1.flat, xy2.flat)]).reshape(xy1.shape)

fig, ax = plt.subplots(figsize=(8, 6))
line_style = dict(levels = [-1.0, 0.0, 1.0],
                  linestyles = ['dashed', 'solid', 'dashed'],
                  colors = 'gray', linewidths=1)
ax.scatter(X[:, 0], X[:, 1], c=y, **point_style)
ax.contour(xy1, xy2, Z, **line_style)


format_plot(ax, 'Modelo aprendido a partir dos dados de entrada')
ax.axis([-1, 4, -2, 7])

#fig.savefig('figures/classificador2.png')
```

Agora que esse modelo foi treinado, ele pode ser generalizado para dados novos e não rotulados. Em outras palavras, podemos pegar um novo conjunto de dados, traçar essa linha de modelo através dele e atribuir rótulos aos novos pontos com base neste modelo. Este estágio geralmente é chamado de previsão. Veja a figura a seguir:

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

ax[0].scatter(X2[:, 0], X2[:, 1], c='gray', **point_style)
ax[0].axis([-1, 4, -2, 7])

ax[1].scatter(X2[:, 0], X2[:, 1], c=y2, **point_style)
ax[1].contour(xy1, xy2, Z, **line_style)
ax[1].axis([-1, 4, -2, 7])

format_plot(ax[0], 'Dados não conhecidos')
format_plot(ax[1], 'Classes preditas')

#fig.savefig('figures/classificador3.png')
```

Essa é a ideia básica de uma tarefa de classificação em aprendizado de máquina, onde "classificação" indica que os dados têm rótulos de classe discretos. À primeira vista, isso pode parecer bastante trivial: seria relativamente fácil simplesmente olhar para esses dados e traçar uma linha discriminatória para realizar essa classificação. Um benefício da abordagem de aprendizado de máquina, no entanto, é que ela pode generalizar para conjuntos de dados muito maiores em muito mais dimensões.

Por exemplo, isso é semelhante à tarefa de detecção automática de spam para e-mail; neste caso, podemos usar os seguintes recursos e rótulos:

- *característica 1, característica 2, etc.* → contagens normalizadas de palavras ou frases importantes ("Viagra", "Príncipe Nigeriano", etc.)

- *classe* → "spam" ou "não é spam"

Para o conjunto de treinamento, esses rótulos podem ser determinados por inspeção individual de uma pequena amostra representativa de emails; para os e-mails restantes, o rótulo seria determinado usando o modelo. Para um algoritmo de classificação adequadamente treinado com recursos bem construídos o suficiente (normalmente milhares ou milhões de palavras ou frases), esse tipo de abordagem pode ser muito eficaz.

# Dividindo o dataset em subconjuntos de treinamento e teste

Para avaliar o desempenho de um classificador, você deve sempre testar o modelo em dados não visualizados. Portanto, antes da construção de um modelo, divida seus dados em duas partes: um conjunto de treinamento e um conjunto de testes.
Você usa o conjunto de testes para treinar e avaliar o modelo durante o estágio de desenvolvimento. Então você usa o modelo treinado para fazer previsões no conjunto de testes não visualizado. Essa abordagem lhe dá uma noção do desempenho e robustez do modelo.
Felizmente, o sklearn tem uma função chamada train_test_split(), que divide seus dados nesses conjuntos. Importe a função e em seguida utilize-a para dividir os dados:

```python
from sklearn.model_selection import train_test_split

# Dividir nossos dados
train, test, train_labels, test_labels = train_test_split(features,
                                                          labels,
                                                          test_size=0.33,
                                                          random_state=42)
```

```python
train_labels
```

A função divide aleatoriamente os dados usando o parâmetro test_size. Neste exemplo, agora temos um conjunto de testes (test) que representa 33% do dataset original. Os dados restantes (train) formam então os dados de treinamento. Também temos os respectivos rótulos para ambas as variáveis train/test, ou seja, train_labels e test_labels.

Os dados de treinamento e teste são amostras da base original e devem respeitar as mesmas discribuições de frequência.

```python
n, bins, patches = plt.hist(iris['data'][setosa,0], density=False, facecolor='g', alpha=0.75, bins=7) # verde para a base completa
n, bins, patches = plt.hist(np.array(train)[train_labels==0,0], density=False, facecolor='b', alpha=0.75, bins=7) # azul para a base de treinamento
n, bins, patches = plt.hist(np.array(test)[test_labels==0,0], density=False, facecolor='r', alpha=0.75, bins=7) # vermelho para a base de teste
intervalo = np.linspace(media-5*desvio,media+5*desvio, num=50)
rv = stat.norm(loc=media, scale=desvio)
plt.plot(intervalo, rv.pdf(intervalo), 'k-', label='pdf')
plt.xlabel('Valores')
plt.ylabel('Frequência')

plt.title('Distribuição normal')
plt.grid(True)
plt.show()
```

## Árvores de Decisão

Existem muitos modelos para machine learning, e cada modelo tem seus pontos fortes e fracos. Vamos começar por um algoritmo simples que geralmente funciona bem em tarefas de classificação, as Árvores de Decisão.

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf = clf.fit(train, train_labels) # o método fit "treina o modelo"
```

Depois de treinarmos o modelo, podemos usar o modelo treinado para fazer previsões no nosso conjunto de teste, o que fazemos utilizando a função predict(). A função predict() retorna uma matriz de previsões para cada instância de dados no conjunto de testes. Podemos então, imprimir nossas previsões para ter uma ideia do que o modelo determinou.

```python
preds = clf.predict(test)
print(preds)
```

Usando a matriz de rótulos de classe verdadeira não utilizada no treinamento, podemos avaliar a precisão dos valores previstos do nosso modelo comparando as duas matrizes (test_labels vs. preds). Utilizaremos a função accuracy_score() do sklearn para determinar a precisão do nosso classificador de machine learning.

```python
from sklearn.metrics import accuracy_score

# Avaliar a precisão
print(accuracy_score(test_labels, preds))
```

Podemos também avaliar os resultados da classificação por meio da matrix de confusão:

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_labels, preds)
```

Note que o classificador se enganou em somente uma instância da base de teste.

Uma das grandes vantagens do uso das árvores de decisão em tarefas de classificação é que elas são intuitivas e sua visualização gráfica dá uma explicação clara das regras utilizadas pelo classificador. No código abaixo mostramos como criar essa visualização usando a biblioteca pydotplus.

```python
from six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

Para leitura da árvore note que a desigualdade referencia uma das variáveis de entrada (X0 a X3) e um ponto de corte. Em seguida, 'samples' indica a quantidade instâncias naquele nó da árvore e 'value' a distribuição entre as classes.

Em certos casos os classificadoes tendem a configurações excessivamente dependentes dos dados fornecidos e generalizam mal em dados de teste. É o chamado <b>overfitting</b>. No caso das árvores de decisão isso ocorre criando ramificações muito profundas, com galhos lidando com subconjuntos muito específicos de instâncias, que inclusive podem ser simples outliers. Para contornar essa situação é possível limitar a profundidade das ramificações, o que causa a chamada "poda" da árvore. Para isso use o parâmetro max_depth.

```python
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)
clf = clf.fit(train, train_labels)
preds = clf.predict(test)
print(accuracy_score(test_labels, preds))
confusion_matrix(test_labels, preds)
```

## Stochastic Gradient Descent (Descendente do gradiente estocástico)

Os algoritmos de aprendizado de máquina usam otimização o tempo todo. Minimizamos a perda, o erro ou maximizamos algum tipo de função de pontuação. A descida de gradiente é o algoritmo de otimização "Hello world", visto provavelmente em qualquer curso de aprendizado de máquina.

Stochastic Gradient Descent (SGD) é uma abordagem simples, mas muito eficiente, para ajustar classificadores lineares e regressores sob funções de perda convexa.

Para compreender o SGD imagine o problema de otimização como uma busca dentro de um espaço de soluções. Nesse espaço cada ponto é um solução possível (uma função), as dimensões do espaço são os hiperparâmetros da função mais o erro que aquela configuração de função apresenta nos dados de treinamento.

![](https://i.stack.imgur.com/4H2vs.png)

A partir de uma ponto de partida a otimização busca uma solução de menor erro, ou seja, ela recalcula os hiperparâmetros da função buscando "descer a ladeira" do espaço de soluções à busca de um ponto mínimo.

O SGD é baseado numa estratégia linear que supõe que localmente o espaço de soluções pode ser descrito por uma função convexa. As funções convexas são bastante simples e supõem uma "direção" única de descida ou subida dos valores de saída da função dadas as suas entradas.

![](https://upload.wikimedia.org/wikipedia/commons/thumb/c/c2/Convex-function-graph-1.png/350px-Convex-function-graph-1.png)

Diversos modelos ditos "lineares" como máquinas de vetor de suporte - (Support Vector Machine - SVM linear)), Perceptron e regressão logística são baseados na descida do gradiente. Embora o SGD já exista na comunidade de aprendizado de máquina há muito tempo, ele recebeu uma quantidade considerável de atenção recentemente no contexto do aprendizado em grande escala devido à sua simplicidade.

SGD foi aplicado com sucesso a problemas de aprendizado de máquina esparsos e em grande escala, freqüentemente encontrados na classificação de textos e processamento de linguagem natural. Dado que os dados são esparsos, os classificadores neste módulo escalam facilmente para problemas com mais de 10^5 exemplos de treinamento e mais de 10^5 dimensões.

Entretanto, em problemas reais de aprendizagem de máquina os espaços não são convexos. O que exige abordagens não lineares como veremos mais adiante.

Estritamente falando, SGD é apenas uma técnica de otimização e não corresponde a uma família específica de modelos de aprendizado de máquina. É apenas uma forma de treinar um modelo. Freqüentemente, uma instância de SGDClassifier ou SGDRegressor terá um estimador equivalente na API scikit-learn, potencialmente usando uma técnica de otimização diferente. Por exemplo, usando SGDClassifier (loss = 'log') resulta em regressão logística, ou seja, um modelo equivalente a LogisticRegression que é ajustado via SGD em vez de ser ajustado por um dos outros solucionadores em LogisticRegression. Da mesma forma, SGDRegressor (loss = 'squared_loss', penalty = 'l2') e Ridge resolvem o mesmo problema de otimização, por meios diferentes.

As vantagens do Estocástico Gradiente Descendente são:

- Eficiência.

- Facilidade de implementação (muitas oportunidades para ajuste de código).

As desvantagens da Descida de Gradiente Estocástico incluem:

- SGD requer vários hiperparâmetros, como o parâmetro de regularização e o número de iterações.

- SGD é sensível às escolhas de inicialização dos hiperparâmetros

- SGD é sensível à escala das dimensões de entrada.


# Support Vector Machines - SVM

O algoritmo de SVM’s se baseia na idéia de que dados dois conjuntos de pontos linearmente separáveis no espaço, ao se encontrar um hiperplano que separe esses dois conjuntos, a classificação de um novo ponto torna-se trivial, pois basta verificar o sinal do resultado ao se inserir esse ponto na equação do hiperplano encontrado.  Como podem existir infinitos hiperplanos que separam dois conjuntos de pontos linearmente separáveis no espaço, o hiperplano cuja margem para os pontos mais próximos (vetores suporte) é a maior é o hiperplano que minimiza o risco de se classificar erroneamente um novo ponto. O desafio do algoritmo de SVM’s é , portanto, o de encontrar o hiperplano com a maior margem para os pontos mais próximos a ele.

![](https://upload.wikimedia.org/wikipedia/commons/3/31/%D7%9E%D7%9B%D7%95%D7%A0%D7%AA_%D7%95%D7%95%D7%A7%D7%98%D7%A8%D7%99%D7%9D_%D7%AA%D7%95%D7%9E%D7%9B%D7%99%D7%9D_%D7%93%D7%95%D7%92%D7%9E%D7%90.jpg)

As vantagens das máquinas do SVM são:

- Eficaz em espaços dimensionais elevados.

- Ainda eficaz nos casos em que o número de dimensões é maior do que o número de amostras.

- Usa um subconjunto de pontos de treinamento na função de decisão (chamados de vetores de suporte), portanto, também é eficiente em termos de memória.

- Versátil: diferentes funções do Kernel podem ser especificadas para a função de decisão. Kernels comuns são fornecidos, mas também é possível especificar kernels personalizados.

As desvantagens das máquinas de vetores de suporte incluem:

- Se o número de dimensões for muito maior do que o número de amostras, evite o sobreajuste ao escolher as funções do Kernel, pois o termo de regularização é crucial.

- Os SVMs não fornecem estimativas de probabilidade diretamente, elas são calculadas usando uma validação cruzada quíntupla cara (consulte Pontuações e probabilidades abaixo).

- Não funciona bem em conjuntos de dados muito grandes, pois exige inversão de matriz - aumentando a complexidade computacional com até o cubo do volume de dados;

- Não funciona bem em conjunto de dados com grande quantidade de ruídos;

- Se as classes estiverem muito sobrepostas deve-se utilizar apenas evidências independentes (devido ao fato de não ser muito bom com dados com muitos ruídos);

O SVM é um exemplo de modelo de kernel. No aprendizado de máquina, as máquinas kernel são uma classe de algoritmos para análise de padrões, cujo membro mais conhecido é o SVM. A tarefa geral da análise de padrões é encontrar e estudar tipos gerais de relações (por exemplo, clusters, classificações, componentes principais, correlações, classificações) em conjuntos de dados. Para muitos algoritmos que resolvem essas tarefas, os dados em representação bruta devem ser explicitamente transformados em representações de vetor de características por meio de um mapa de características especificado pelo usuário. Em contraste, os métodos do kernel requerem apenas um kernel especificado pelo usuário, ou seja, uma função de similaridade sobre pares de pontos de dados na representação bruta. Normalmente um kernel é uma função que a similaridade baseada na distância entre os os pontos. Quanto mais longe menos similares. Um kernel comumente utilizado em ML é o RBF: Radial Basis Function, baseado na distância euclidiana.

```python
print(__doc__)

import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets


def make_meshgrid(x, y, h=.02):
    """Create a mesh of points to plot in

    Parameters
    ----------
    x: data to base x-axis meshgrid on
    y: data to base y-axis meshgrid on
    h: stepsize for meshgrid, optional

    Returns
    -------
    xx, yy : ndarray
    """
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy


def plot_contours(ax, clf, xx, yy, **params):
    """Plot the decision boundaries for a classifier.

    Parameters
    ----------
    ax: matplotlib axes object
    clf: a classifier
    xx: meshgrid ndarray
    yy: meshgrid ndarray
    params: dictionary of params to pass to contourf, optional
    """
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out


# import some data to play with
iris = datasets.load_iris()
# Take the first two features. We could avoid this by using a two-dim dataset
X = iris.data[:, :2]
y = iris.target

# we create an instance of SVM and fit out data. We do not scale our
# data since we want to plot the support vectors
C = 1.0  # SVM regularization parameter
models = (svm.SVC(kernel='linear', C=C),
          svm.LinearSVC(C=C, max_iter=10000))
models = (clf.fit(X, y) for clf in models)

# title for the plots
titles = ('SVC with linear kernel',
          'LinearSVC (linear kernel)')

# Set-up 2x2 grid for plotting.
fig, sub = plt.subplots(1, 2,figsize=(12,5))
plt.subplots_adjust(wspace=0.4, hspace=0.4)

X0, X1 = X[:, 0], X[:, 1]
xx, yy = make_meshgrid(X0, X1)

for clf, title, ax in zip(models, titles, sub.flatten()):
    plot_contours(ax, clf, xx, yy,
                  cmap=plt.cm.coolwarm, alpha=0.8)
    ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xlabel('Sepal length')
    ax.set_ylabel('Sepal width')
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)

plt.show()
```

# Multilayer Perceptron

Multi-layer Perceptron (MLP) é um algoritmo de aprendizado supervisionado que aprende uma função de R^m -> R^n por treinamento em um conjunto de dados, onde é m o número de dimensões para entrada e n é o número de dimensões para saída.

Dado um conjunto de entradas X = [x1,x2,..., xm] e um alvo y , ele pode aprender um aproximador de função não linear para classificação ou regressão. É diferente da regressão logística, pois entre a camada de entrada e a de saída pode haver uma ou mais camadas não lineares, chamadas camadas ocultas. A Figura abaixo mostra um MLP de uma camada oculta com saída escalar.

![](https://i0.wp.com/spotintelligence.com/wp-content/uploads/2024/02/multilayer-perceptron-architecture-1024x576.webp?resize=1024%2C576&ssl=1)


As vantagens do Perceptron multicamadas são:

- Capacidade de aprender modelos não lineares.

- Capacidade de aprender modelos em tempo real (aprendizado on-line) usando o partial_fit.

As desvantagens do Multi-layer Perceptron (MLP) incluem:

- O MLP com camadas ocultas tem uma função de perda não convexa onde existe mais de um mínimo local. Portanto, inicializações de peso aleatório diferentes podem levar a uma precisão de validação diferente.

- O MLP requer o ajuste de vários hiperparâmetros, como o número de neurônios ocultos, camadas e iterações.

- O MLP é sensível à escala das dimensões.

## O que é possível (necessário) definir na Rede Neural

- Camadas "escondidadas" (hidden_layers): tupla com o número de neurônios por camada escondida (comprimento da tupla define o número de camadas)

- Função de ativação: activation {‘identity’, ‘logistic’, ‘tanh’, ‘relu’}, default=’relu’

 - Função de ativação para a camada oculta.

  - 'identity', ativação sem operação, retorna f (x) = x

  - 'logistic', a função sigmóide logística, retorna f (x) = 1 / (1 + exp (-x)).

  - ‘tanh’, a função tan hiperbólica, retorna f (x) = tanh (x).

  - ‘relu’, a função da unidade linear retificada, retorna f (x) = max (0, x)


- Otimizador de custo : solver{‘lbfgs’, ‘sgd’, ‘adam’}, default=’adam’.

 - 'Sgd' a 'adam' são baseados na descida do gradiente estocático. Como usam derivadas parciais (matrizes chamadas Jacobianas) são chamados de métodos "newtonianos"

 - ‘Lbfgs’ é um otimizador na família de métodos quasi-newtonianos (aproxima a Jacobiana)

 -  O otimizador de custo é o orquestrador da minimização de perdas da rede neural. Para o conjunto de dados, o objetivo de otimização é a redução do erro médio em todas as instâncias de dados em todo o conjunto de dados. O otimizador define como as instâncias são apresentadas à rede (forward) e como a correção dos pesos é feita (backpropagation).

 - Nota: O solucionador padrão ‘adam’ funciona muito bem em conjuntos de dados relativamente grandes (com milhares de amostras de treinamento ou mais) em termos de tempo de treinamento e pontuação de validação. Para pequenos conjuntos de dados, no entanto, 'lbfgs' pode convergir mais rápido e ter um desempenho melhor.

- Termo de regularização - penalidade L2: alpha, default = 0.0001

- Numéros máximo de iterações (max_iter): por padrão 200

- Tamanho do lote usado em cada iteração da descida do gradiente: por padrão (opção 'auto') batch_size=min(200, n_samples)

- Tolerância para a otimização: tol, default=1e-4

 - Quando a perda ou pontuação não está melhorando em pelo menos tol para para um certo número de iterações sucessivas (definido por n_iter_no_change) iterações consecutivas a convergência é considerada atingida e o treinamento para.

- Parada antecipada: early_stopping, default=False

 - Se usar a parada antecipada encerra o treinamento quando a pontuação de validação não estiver melhorando. Se definido como verdadeiro, ele reservará automaticamente 10% dos dados de treinamento como validação e encerrará o treinamento quando a pontuação de validação não estiver melhorando em pelo menos tol para n_iter_no_change períodos consecutivos.  Eficaz apenas quando solver = ’sgd’ ou ‘adam’ (descida de gradiente estocástica).

- Cronograma de taxa de aprendizagem para atualizações dos  pesos : learning_rate {‘constant’, ‘invscaling’, ‘adaptive’}, default=’constant’. Usado somente com o solver='sgd'.

 - 'constant'é uma taxa de aprendizagem constante dada por' learning_rate_init '.

 - ‘invscaling’ diminui gradualmente a taxa de aprendizagem em cada passo de tempo ‘t’ usando um expoente de escala inversa de ‘power_t’. Effective_learning_rate = learning_rate_init / pow (t, power_t)

 - ‘adaptative’ mantém a taxa de aprendizagem constante para ‘learning_rate_init’ enquanto a perda de treinamento continua diminuindo. Cada vez que duas épocas consecutivas falham em diminuir a perda de treinamento em pelo menos tol, ou falham em aumentar a pontuação de validação em pelo menos tol se ‘early_stopping’ estiver ativado, a taxa de aprendizado atual é dividida por 5

# Ensembles

O objetivo dos métodos de ensemble é combinar as previsões de vários estimadores de base construídos com um determinado algoritmo de aprendizagem, a fim de melhorar a generalização / robustez sobre um único estimador.

Duas famílias de métodos de conjunto são geralmente distinguidas:

Nos métodos de cálculo da média, o princípio orientador é construir vários estimadores de forma independente e, em seguida, calcular a média de suas previsões. Em média, o estimador combinado é geralmente melhor do que qualquer um do estimador de base única porque sua variância é reduzida.

Exemplos: Bagging methods, Forests of randomized trees (florestas de árvores aleatórias), ...

Em contraste, nos métodos de boosting, estimadores de base são construídos sequencialmente e tenta-se reduzir o viés do estimador combinado. A motivação é combinar vários modelos fracos para produzir um conjunto poderoso.

Exemplos: AdaBoost, Gradient Tree Boosting, ...

Floresta Aleatória (Random Forest) é um algoritmo de aprendizagem de máquina flexível e fácil de usar que produz excelentes resultados a maioria das vezes, mesmo sem ajuste de hiperparâmetros. É também um dos algoritmos mais utilizados, devido à sua simplicidade e o fato de que pode ser utilizado para tarefas de classificação e também de regressão.

A “floresta” que ele cria é uma combinação (ensemble) de árvores de decisão, na maioria dos casos treinados com o método de bagging. A idéia principal do método de bagging é que a combinação dos modelos de aprendizado aumenta o resultado geral.

No Random Forest cada árvore no conjunto é construída a partir de uma amostra retirada com substituição (ou seja, uma amostra de bootstrap) do conjunto de treinamento.

Além disso, ao dividir cada nó durante a construção de uma árvore, a melhor divisão é encontrada em todos os recursos de entrada ou em um subconjunto aleatório de tamanho max_features.

O objetivo dessas duas fontes de aleatoriedade é diminuir a variância do estimador da árvore de decisão. Na verdade, as árvores de decisão individuais geralmente exibem alta variação e tendem a se ajustar demais. A aleatoriedade injetada nas florestas produz árvores de decisão com erros de predição um tanto dissociados. Ao tirar uma média dessas previsões, alguns erros podem ser cancelados. As florestas aleatórias alcançam uma variação reduzida combinando diversas árvores, às vezes ao custo de um ligeiro aumento no viés. Na prática, a redução da variância é freqüentemente significativa, resultando em um modelo geral melhor.

# Aprendizagem não supervisionada

Os métodos de aprendizado não supervisionado são guiados exclusivamente pela estrutura intrínseca dos dados e/ou seus padrões de similaridade ao longo do processo de aprendizagem. Ou seja, algoritmos de aprendizado não supervisionado não utilizam qualquer informação sobre classes ou variáveis-alvo, tal como ocorre no paradigma de aprendizado supervisionado.



![Comparação entre uma tarefa supervisionada (à esquerda) e não supervisionada (à direita)](https://techdifferences.com/wp-content/uploads/2018/01/Untitled.jpg)



Existem diversas tarefas que integram o aprendizado não supervisionado:

1. **Clustering (ou agrupamento)** [vamos ver nesta aula]: esse é a tarefa mais comum em aprendizagem não supervisionado. É usada para encontrar padrões e estrutura nos dados por meio de agrupamentos. Pares de dados classificados em um mesmo cluster são mais similares do que pares de dados classificados em clusters distintos. Os clusters são modelados usando uma medida de similaridade que é definida por meio de métricas de distância, tal como a distância Euclidiana, cosseno ou alguma outra variante. Existem diversos algoritmos dessa tarefa que serão explorados nesta aula.

2. **Redução da dimensionalidade** [já discutida em aulas anteriores]: essa tarefa consiste na transformação de dados de um espaço de alta dimensão em um espaço de baixa dimensão, de tal forma que a representação de baixa dimensão retenha algumas propriedades significativas dos dados originais. Um princípio subjacente a técnicas pertencentes a essa tarefa é que "dados similares devem continuar próximos e dados diferentes devem continuar distantes uns dos outros". A manutenção dessa propriedade é importante para que a característica topológica dos dados não seja perdida durante a transformação. A redução da dimensionalidade é utilizada quando temos poucos dados em uma dimensão alta demais, de tal forma a evitar a "maldição da dimensionalidade". Quando os dados estão esparsos demais (o que ocorre em dimensões muito altas se tivermos poucos dados), os algoritmos tradicionais de aprendizado de máquina não desempenham bem. Assim, a redução da dimensionalidade dos dados é um bom passo de pré-processamento.

3. **Regras de associação** [vamos ver em aula posterior]: essa tarefa permite o estabelecimento de associações entre subconjuntos de dados. Por exemplo, pessoas que compram uma nova casa provavelmente comprarão móveis novos. A descoberta dessas associações é uma tarefa bem comum no meio corporativo e bem utilizada em *e-commerce*.

## K-Means

O algoritmo K-Means segue os seguintes passos:

1. Na primeira etapa, chamada de etapa de inicialização, o usuário escolhe um número de clusters. Cada cluster é representado por um centróide. O K-Means dispõe os centróides iniciais de forma aleatória na base de dados (algumas variantes do K-Means possuem outras estratégias).

2. Na segunda etapa, o K-Means designa para cada dado o centróide mais próximo a ele. Cada centróide representa um cluster no algoritmo.

3. Na terceira etapa, o K-Means atualiza a localização física do centróide usando o valor médio de todas os dados atribuídos ao grupo correspondente àquele centróide.

4. A diferença entre o posicionamento novo e antigo de cada centróide é calculada. Se essa diferença for não pequena (até um limite estabelecido pelo usuário), volta-se para o Passo 2. Se a diferença for pequena o suficiente, significa que os centróides convergiram e o algoritmo termina. O cluster a que cada dado pertence é determinado pelo centróide mais próximo.

![Funcionamento do K-Means (iterações)](https://drive.google.com/uc?id=14Z31COYabi4FIT_F9cIDDOwpp6YfZNga)

Para um número de iterações suficientemente grande, o K-means sempre convergirá. No entanto, a convergência pode ocorrer em um mínimo local. O vetor de convergência depende da inicialização dos centróides. Um método para ajudar a resolver esse problema é o esquema de inicialização `k-means++`, que foi implementado no `scikit-learn` (use o parâmetro `init = 'k-means++'`). Isso inicializa os centróides distantes uns dos outros, aumentando as chances de obtermos resultados melhores do que a inicialização aleatória.

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
import seaborn as sns

# Gera dados em forma circulares em duas dimensões, com quatro clusters
X, y = make_blobs(n_samples=1000, n_features=2, centers=4, random_state=42)

synthetic_data = pd.DataFrame(np.c_[X, y], columns=['x1', 'x2', 'y'])

sns.relplot(data = synthetic_data,
            x = 'x1',
            y = 'x2',
            hue = 'y')
```

Vamos criar um agrupador `KMeans` no `sklearn` com 4 clusters, usando o método de inicialização `k-means++` e rodando dez vezes consecutivas de forma independente. O uso desse método de inicialização e as execuções independentes servem para minimizarmos o problema de atingirmos um mínimo local. O agrupamento que é retornado pelo `KMeans` é aquele em que há um melhor fit dos dados.

```python
NUMBER_OF_CLUSTERS = 4
kmeans = KMeans(n_clusters = NUMBER_OF_CLUSTERS, init = 'k-means++', max_iter = 300, n_init = 10,random_state = 0)
# n_init define quantas tentativas de centróides iniciais são feitas
```

O fitting do modelo utiliza a função `fit`, que é similar ao que estávamos utilizando para classificação. A diferença aqui é que não passamos mais as classes do problema `y`.

```python
y_kmeans = kmeans.predict(X)
```

Disposição dos clusters no espaço e os centróides estimados de cada cluster:

**IMPORTANTE**: Nem sempre sabemos a quantidade de clusters ex-ante ao problema. Nesses casos, podemos usar o **elbow method**, que consiste em testar vários valores de clusters e encontrar o "cotovelo" do gráfico representando o ponto em que o erro para de diminuir muito em relação aos pontos anteriores.

```python
from yellowbrick.cluster import KElbowVisualizer

kmeans = KMeans(init = 'k-means++', max_iter = 300, n_init = 10,random_state = 0)

visualizer = KElbowVisualizer(kmeans, k=(1,12))

visualizer.fit(X)        # Fit the data to the visualizer
```

Outra forma de avaliar a quantidade adequada de clusters é por meio da redução da chamada "inércia". A inércia pode ser reconhecida como uma medida de quão internamente os clusters são coerentes.  A inércia é o somatório do erro quadrático de cada cluster. Ou seja, quanto menor a inércia mais denso é o cluster (mais próximos uns dos outros os pontos do cluster estão).

Embora seja uma informação útil ela sofre de várias desvantagens:

- A inércia pressupõe que os aglomerados são convexos e isotrópicos, o que nem sempre é o caso. Ele responde mal a aglomerados alongados ou variedades com formas irregulares.

- A inércia não é uma métrica normalizada: apenas sabemos que valores mais baixos são melhores e zero é o ideal. Mas em espaços de dimensões muito altas, as distâncias euclidianas tendem a se tornar infladas (este é um exemplo da chamada “maldição da dimensionalidade”). A execução de um algoritmo de redução de dimensionalidade, como a análise de componente principal (PCA) antes do agrupamento k-means pode aliviar esse problema e acelerar os cálculos.

```python
ks = range(1, 30)
inertias = []

for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k,random_state = 0)

    # Fit model to samples
    model.fit(X)

    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
```

```python
# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

```python
def benchmark(clf):
    print('_' * 80)
    print("Training: ")
    print(clf)
    t0 = time()
    clf.fit(X_train, y_train)
    train_time = time() - t0
    print("train time: %0.3fs" % train_time)

    t0 = time()
    pred = clf.predict(X_test)
    test_time = time() - t0
    print("test time:  %0.3fs" % test_time)

    score = metrics.accuracy_score(y_test, pred)
    print("accuracy:   %0.3f" % score)

    if hasattr(clf, 'coef_'):
        print("dimensionality: %d" % clf.coef_.shape[1])
        print("density: %f" % density(clf.coef_))

    #print("classification report:")
    #print(metrics.classification_report(y_test, pred,
    #                                    target_names=target_names))

    print("confusion matrix:")
    print(metrics.confusion_matrix(y_test, pred))

    print()
    clf_descr = str(clf).split('(')[0]
    return clf_descr, score, train_time, test_time
```

```python
print(__doc__)

# https://scikit-learn.org/stable/auto_examples/classification/plot_classifier_comparison.html
# Code source: Gaël Varoquaux
#              Andreas Müller
# Modified for documentation by Jaques Grobler
# License: BSD 3 clause

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

h = .02  # step size in the mesh

names = ["Nearest Neighbors", "Linear SVM", "RBF SVM", "Gaussian Process",
         "Decision Tree", "Random Forest", "Neural Net", "AdaBoost",
         "Naive Bayes", "QDA"]

classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="linear", C=0.025),
    SVC(gamma=2, C=1),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis()]

X, y = make_classification(n_features=2, n_redundant=0, n_informative=2,
                           random_state=1, n_clusters_per_class=1)
rng = np.random.RandomState(2)
X += 2 * rng.uniform(size=X.shape)
linearly_separable = (X, y)

datasets = [make_moons(noise=0.3, random_state=0),
            make_circles(noise=0.2, factor=0.5, random_state=1),
            linearly_separable
            ]

figure = plt.figure(figsize=(27, 9))
i = 1
# iterate over datasets
for ds_cnt, ds in enumerate(datasets):
    # preprocess dataset, split into training and test part
    X, y = ds
    X = StandardScaler().fit_transform(X)
    X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=.4, random_state=42)

    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    # just plot the dataset first
    cm = plt.cm.RdBu
    cm_bright = ListedColormap(['#FF0000', '#0000FF'])
    ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
    if ds_cnt == 0:
        ax.set_title("Input data")
    # Plot the training points
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright,
               edgecolors='k')
    # Plot the testing points
    ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6,
               edgecolors='k')
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xticks(())
    ax.set_yticks(())
    i += 1

    # iterate over classifiers
    for name, clf in zip(names, classifiers):
        ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)

        # Plot the decision boundary. For that, we will assign a color to each
        # point in the mesh [x_min, x_max]x[y_min, y_max].
        if hasattr(clf, "decision_function"):
            Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
        else:
            Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]

        # Put the result into a color plot
        Z = Z.reshape(xx.shape)
        ax.contourf(xx, yy, Z, cmap=cm, alpha=.8)

        # Plot the training points
        ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright,
                   edgecolors='k')
        # Plot the testing points
        ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright,
                   edgecolors='k', alpha=0.6)

        ax.set_xlim(xx.min(), xx.max())
        ax.set_ylim(yy.min(), yy.max())
        ax.set_xticks(())
        ax.set_yticks(())
        if ds_cnt == 0:
            ax.set_title(name)
        ax.text(xx.max() - .3, yy.min() + .3, ('%.2f' % score).lstrip('0'),
                size=15, horizontalalignment='right')
        i += 1

plt.tight_layout()
plt.show()
```
