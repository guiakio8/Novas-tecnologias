## Group by

Permite a agregação de dados em segmentos.

Tem três partes:

- qual a coluna cujos valores vão ser agregados (gera uma linha por valor dessa coluna e ela vai ser o index da resposta)

- qual a coluna da qual os valores vão ser totalizados (vai gerar um valor por linha do index)

- qual a operação de agragação vai ser feita sobre a coluna totalizada

```python
# agrupa por estado (index) e na coluna de total de casos por 100 mil habitantes obtém o valor máximo
# ou seja, obtém o maior valor diário da série por estado do total de casos por 100 mil habitantes
seriecovid.groupby('state')['totalCases_per_100k_inhabitants'].max()
```
